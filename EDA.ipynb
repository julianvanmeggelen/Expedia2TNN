## Create embedding id's

```python
from collections import defaultdict
maps_to_ind = {}
maps_from_ind = {}
for col in cat_feature_col:
    unique = np.sort(df_train[col].unique().astype('int'))
    print(col+'_'*50, '\n',unique)
    nElements = len(unique+1)
    map_from_ind = {i+1:_ for i,_ in enumerate(unique)}
    map_to_ind = {_:i+1 for i,_ in enumerate(unique)}
    maps_from_ind[col] = defaultdict(lambda:None,map_from_ind)
    maps_to_ind[col] = defaultdict(lambda:0,map_to_ind)
```
